```python
from keras.datasets import mnist
from keras_preprocessing.image import ImageDataGenerator
from matplotlib import pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print("Train X and y shape:", X_train.shape, y_train.shape)
print("Test X and y shape:", X_test.shape, y_test.shape)
```

```python
for i in range(9):
    plt.subplot(330+1+i)
    plt.imshow(X_train[i], cmap=plt.get_cmap('gray'))
plt.show()
```

```python
#reshape and convert from int to float
Xtrain = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32')
Xtest = X_test.reshape(X_test.shape[0], 28, 28, 1).astype('float32')
```

```python
# define data preparation
datagen = ImageDataGenerator(featurewise_center=True, featurewise_std_normalization=True)
```

```python
# fit parameters from data
datagen.fit(Xtrain)
```

```python
# configure batch size and retrieve one batch of images
for (Xbatch, ybatch) in datagen.flow(Xtrain, y_train, batch_size=9):
    for i in range(9):
        plt.subplot(330+1+i)
        plt.imshow(Xbatch[i].reshape(28,28), cmap=plt.get_cmap('gray'))
    plt.show()
    break
```

# ZCA whitening

```python
datagen2 = ImageDataGenerator(zca_whitening=True)
datagen2.fit(Xtrain)
```

```python
# configure batch size and retrieve one batch of images
for (Xbatch, ybatch) in datagen2.flow(Xtrain, y_train, batch_size=9):
    for i in range(9):
        plt.subplot(330+1+i)
        plt.imshow(Xbatch[i].reshape(28,28), cmap=plt.get_cmap('gray'))
    plt.show()
    break
```
